Прочитайте файл, преобразуйте индекс "Дата" в DatetimeIndex и постройте график изменения температуры за указанный период.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('temperature.csv', sep=';', encoding='utf-8-sig', parse_dates=['Дата'], index_col='Дата')
plt.figure(figsize=(8, 5))
plt.plot(df.index, df['Температура'], marker='o', label='Температура')
plt.title("Изменение температуры")
plt.xlabel("Дата")
plt.ylabel("Температура (°C)")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
```
